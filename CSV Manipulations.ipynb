```python
import os
import glob
import pandas as pd
os.chdir("output")
```

```python
extension = "csv"
all_filenames = [i for i in glob.glob("*.{}".format(extension))]
```

```python
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames])
combined_csv.reset_index(drop=True, inplace=True)
combined_csv["Lenguaje"] = combined_csv["Actividad"].apply(lambda x: "Python" if "Python" in x else "JS")
combined_csv["Intentada?"] = combined_csv["Descripción"].apply(lambda x: "Nel" if "Not Submitted" in x else "Simón")
combined_csv["Correcta?"] = combined_csv["Descripción"].apply(lambda x: "Nel" if "failed" in x else "Simón")
combined_csv = combined_csv[["Actividad", "Usuario", "Lenguaje", "Intentada?", "Correcta?", "Descripción"]]
```

```python
result = pd.DataFrame(
    {
        "Act. Correctas": combined_csv.groupby(["Usuario"])["Correcta?"].apply(lambda x: x.str.contains("Si").sum()),
        "Act. Intentadas": combined_csv.groupby(["Usuario"])["Intentada?"].apply(lambda x: x.str.contains("Si").sum()),
        "Act. con Python": combined_csv.groupby(["Usuario"])["Lenguaje"].apply(lambda x: x.str.contains("Python").sum()),
        "Act. con JS": combined_csv.groupby(["Usuario"])["Lenguaje"].apply(lambda x: x.str.contains("JS").sum()),
    }
).reset_index()
```
